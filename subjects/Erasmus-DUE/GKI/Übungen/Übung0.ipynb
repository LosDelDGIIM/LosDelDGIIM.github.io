## Exercise 11. Functions

1. Write a function `add(a, b)` that returns the sum of two numbers `a` and `b`.
2. Test the function by calling it with different arguments and print the result to the console.
3. Write a function `is_palindrome(word)` that checks if a given word is a palindrome (reads the same forwards and backwards).
4. Test the function with different words and print the result to the console.
5. Write a function `check_weights(weights)` that takes a list of fruits and prints whether each weight $\geq$ 60 or less $\leq$ 60.

```python
def add(a, b):
    return a + b
print(add(5, 3))
print(add(-2, 10))

def is_palindrome(word):
    return word == word[::-1]
print(is_palindrome("radar"))
print(is_palindrome("hello"))

def check_weights(weights):
    for fruit, weight in weights.items():
        if weight >= 60:
            print(f"{fruit} is heavy (weight ≥ 60g).")
        else:
            print(f"{fruit} is light (weight < 60g).")
fruit_weights = {
    'Apple': 150,
    'Banana': 120,
    'Cherry': 10
}
check_weights(fruit_weights)
```

## Exercise 12. Classes

1. Define class `Fruit` with attributes `name`, `weight`, `size`, and `color`.
2. Write a constructor method `__init__(self, name, weight, size, color)` to initialize the attributes.
3. Add a method `print_info()` to the class that prints all the attributes of the fruit.
4. Create an instance of the `Fruit` class with the information of a fruit.
5. Create the instances of the `Fruit` class `orange`, `apple`, `watermelon` with the information of a fruit.
6. Create the class `fruit_salad` which contains the fruits above and has the function `make_salad(fruit_list)`. This function should make a summation over the weights of the used fruits and print a list of the used fruits.
7. Call the `print_info()` method to print the information of the fruit.

```python
class Fruit:
    def __init__(self, name, weight, size, color):
        self.name = name
        self.weight = weight
        self.size = size
        self.color = color
    
    def __str__(self):
        return f"Fruit(name={self.name}, weight={self.weight}, size={self.size}, color={self.color})"
    
    def print_info(self):
        print(f"- Name: {self.name}")
        print(f"- Weight: {self.weight}g")
        print(f"- Size: {self.size}")
        print(f"- Color: {self.color}")
apple = Fruit('Apple', 150, 'Medium', 'Red')
orange = Fruit('Orange', 120, 'Medium', 'Orange')
watermelon = Fruit('Watermelon', 2000, 'Large', 'Green')
print(apple)
print(orange)
print(watermelon)
apple.print_info()


class Fruit_Salad:
    def __init__(self, fruits):
        self.fruits = fruits
    
    def total_weight(self):
        return sum(fruit.weight for fruit in self.fruits)
    
    def make_salad(self):
        print("Making fruit salad with:")
        for fruit in self.fruits:
            print(f"- {fruit.name}")
        print(f"Total weight: {self.total_weight()}g")

print()
salad = Fruit_Salad([apple, orange, watermelon])
salad.make_salad()
```
