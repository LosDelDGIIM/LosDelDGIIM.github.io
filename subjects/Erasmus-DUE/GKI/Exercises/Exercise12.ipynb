```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer, AutoTokenizer

# Loading the model and tokenizer
# Here the GPT-2 model is used
# You can also use other models
# e.g. GPT-2 Medium, Large, etc.
# Note that larger models require more resources
# and may not run on all systems
model_name = "gpt2-medium"
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Configurar el token de padding correctamente
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

eos_token_id = tokenizer.eos_token_id

def process_input(input_text, max_length=70):
    """
    Process the input text and generate a response using the model.
    Args:
        input_text (str): The input text to process.
                            It should be in English and not longer than 512 tokens.
        max_length (int): The maximum length of the generated response.
    Returns:
        str: The generated response from the model.
    """
    # Tokenize the text
    inputs = tokenizer(
        input_text, 
        return_tensors='pt',
        padding=True, 
        return_attention_mask=True
    )

    # Apply the model
    output = model.generate(
        input_ids=inputs['input_ids'],
        attention_mask=inputs['attention_mask'],
        pad_token_id=tokenizer.pad_token_id,
        eos_token_id=eos_token_id,
        max_length=max_length
    )

    # Decode the output
    output_text = tokenizer.decode(output[0], skip_special_tokens=True)
    return output_text
```

### Test generating using parameters

```python
# Control text generation with parameters

def generate_text_with_params(input_text, max_length=50, temperature=0.7, top_k=50):
    inputs = tokenizer(input_text, return_tensors='pt',
                                padding=True, return_attention_mask=True)
    
    output = model.generate(
        
        input_ids=inputs['input_ids'],
        attention_mask=inputs['attention_mask'],
        pad_token_id=tokenizer.pad_token_id,
        eos_token_id=eos_token_id,
        max_length=max_length,
        num_return_sequences=1,
        temperature=temperature,
        top_k=top_k,
        do_sample=True,
        no_repeat_ngram_size=3,
        repetition_penalty=1.2
    )
    
    output_text = tokenizer.decode(output[0], skip_special_tokens=True)
    return output_text

# Test mit den vorgegebenen Parametern # Test with the given parameters
input_text = "Tell me a short story."
response = generate_text_with_params(input_text)
print(f"Input: {input_text}")
print(f"Answer (temperature=0.7, top_k=50): {response}")
```

### Test Different Models

```python
# Comparison of different GPT-2 model sizes

def compare_models(input_text, model_sizes=["gpt2", "gpt2-medium", "gpt2-large"]):
    results = {}
    
    for size in model_sizes:
        print(f"Loading model {size}...")
        try:
            temp_model = GPT2LMHeadModel.from_pretrained(size)
            temp_tokenizer = AutoTokenizer.from_pretrained(size)
            
            if temp_tokenizer.pad_token_id is None:
                temp_tokenizer.pad_token_id = temp_tokenizer.eos_token_id
                
            inputs = temp_tokenizer(input_text, return_tensors='pt',
                                           padding=True, return_attention_mask=True)
            
            output = temp_model.generate(
                input_ids=inputs['input_ids'],
                attention_mask=inputs['attention_mask'],
                pad_token_id=temp_tokenizer.pad_token_id,
                eos_token_id=temp_tokenizer.eos_token_id,
                max_length=50,
                do_sample=True,
                num_return_sequences=1,
                temperature=0.7,
                top_k=50,
                no_repeat_ngram_size=3,
                repetition_penalty=1.2
            )
            
            output_text = temp_tokenizer.decode(output[0], skip_special_tokens=True)
            results[size] = output_text
        except Exception as e:
            results[size] = f"Error loading model: {str(e)}"
    
    return results

# Comparison of the models
input_text = "Tell me a short story."
model_responses = compare_models(input_text, ["gpt2", "gpt2-medium", "gpt2-large"])

for model_name, response in model_responses.items():
    print(f"Model: {model_name}")
    print(f"Response: {response}")
    print("-" * 80)
```

### Test Different Parameters

```python
# Experimenting with temperature and top_k parameters
# Here you can try different values for temperature and top_k

input_text = "The weather is very"
temperatures = [0.01, 0.1, 0.5, 1.0]
top_ks = [1, 5, 20, 100]

# Teste verschiedene temperature-Werte  # Test different temperature values
print("EXPERIMENTING WITH TEMPERATURE VALUES")
print("=" * 80)
for temp in temperatures:
    response = generate_text_with_params(input_text, temperature=temp, top_k=50)
    print(f"temperature = {temp}")
    print(f"Response: {response}")
    print("-" * 80)

# Teste verschiedene top_k-Werte # Test different top_k values
print("\nEXPERIMENTING WITH TOP_K VALUES")
print("=" * 80)
for k in top_ks:
    response = generate_text_with_params(input_text, temperature=0.7, top_k=k)
    print(f"top_k = {k}")
    print(f"Response: {response}")
    print("-" * 80)
```
